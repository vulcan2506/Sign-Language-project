```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def pre_process(text):
    text = text.lower()
    arr = list(text)
```

```python
sign_language_mapping = {
"a":"A1.jpg",
"b":"B3.jpg",    
"c":"B1.jpg"    
}
```

```python
text = "Hii"
pre_process(text)
```

```python
def translate(text):
    sign_language_translation = []
    for char in text:
        if char in sign_language_mapping:
            image_path = sign_language_mapping[char]
            sign_language_translation.append(image_path)
        else:
            sign_language_translation.append(None)

    return sign_language_translation
    
```

```python
input_text = "ab"
sign_language_images = translate(input_text)

for char,image_path in zip(input_text,sign_language_images):
     print(f"Character: {char}, Sign Language Image: {image_path}")
     image = mpimg.imread(image_path)
     plt.imshow(image)
     plt.title(f"Character: {char}")
     plt.axis('off')
     plt.figure(figsize=(2,2))
     plt.show()
     
```
